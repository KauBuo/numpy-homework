#### [Numpy数组创建: 4. 绘制两种随机量对应的折线图](https://www.wolai.com/jt9vUU4FMSXHp9h4NdmRm7#sbEzwddvqqELZExFom79p2) 

```python
#引入合适的模块
import numpy as np
import matplotlib.pyplot as plt
# 生成[0,50]范围内100个数字
x = np.linspace(0, 50, 100)
# 生成100个标准正态分布的随机数数组
y1 = np.random.randn(100)
# 生成100个[0, 1)区间内的均匀分布的随机数
y2 = np.random.rand(100)
# 把两个随机数叠加绘制出来
y3= y1 + y2
plt.plot(x,y3)
plt.plot(x,y2)
plt.show()
```

#### [Numpy数组创建: 6. 给局部狗图添加噪音](https://www.wolai.com/jt9vUU4FMSXHp9h4NdmRm7#gQZCXSkJ2SkxYtC8zuEeom) 

```python
# 引入合适的模块
import cv2
import numpy as np
import matplotlib.pyplot as plt
# 读取灰度图
img = cv2.imread("dog.jpg", 0)
# 截取局部图像
j = img[500:1100, 100:600]
# 获取图像尺寸
h, w = j.shape[:2]
# 生成正态分布随机值数组，并将每个值扩大20倍，作为噪声
noise = np.random.normal(0, 1, (h, w)) * 20
# 将噪声叠加到原图上
noisy_j = cv2.add(j, noise.astype(np.uint8))
# 将处理后的局部图像重新放回原图中
img[500:1100, 100:600] = noisy_j
# 显示处理后的图像
plt.imshow(noisy_j, cmap='gray') 
plt.show()
```
